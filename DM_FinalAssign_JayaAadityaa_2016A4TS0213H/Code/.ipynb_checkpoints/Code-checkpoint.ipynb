# SPADE ALGORITHM ON ONLINE RETAIL DATASET 

## 1. Initialization

```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import copy


data = pd.read_csv('../Data/Online_Retail.csv')
df = pd.DataFrame(data)
df
```

## 2. Pre-Processing

### 2.1. Data Cleaning

```python
#'CustomerID' values are important as they deterimne sequence number. Hence any instance without a 'CustomerID' value 
# is removed. 
df['CustomerID'].replace('', np.nan, inplace=True)
df.dropna(subset=['CustomerID'], inplace=True)
```

### 2.2. Data Reduction

```python
# The attributes which are irrelevant to the current task at hand are dropped
df = df.drop(columns=['Quantity','UnitPrice','Country']) 
df = df[['CustomerID','InvoiceNo','StockCode','Description','InvoiceDate']]
df = df.reset_index()
df = df.drop(columns=['index'])
df
```

### 2.3.1. Data Transformation - 'Cancellation'

```python
# A cancellation order has 'C' appended to its invoice number. To enable later sorting with invoice number,
# 'C' is removed from the invoice number and 'CN' is inserted in front of the Stock Code
df_size = len(df)
i = 0
INNew = [] #The new 'InvoiceNo' attribute
SCNew = [] #The new 'StockCode' attribute
while i<df_size:
    IN = df.loc[i,'InvoiceNo']
    SC = df.loc[i,'StockCode']
    if IN[0] == 'C': #If 'C' is the first element of the 'InvoiceNo', then te modification occurs
        IN = IN[1:]
        SC = 'CN' + SC
    INNew.append(IN)
    SCNew.append(SC)
    i = i+1
i = 0
df.drop(columns=['StockCode','InvoiceNo'])
df['StockCode'] = SCNew
df['InvoiceNo'] = INNew
```

### 2.3.2. Data Transformation - 'Time'

```python
# The objective of this block is to convert the dd/mm/yy H/M format in 'InvoiceDate' attribute to a relative format 
# with respect to the first transaction in terms of days. 
df_size = len(df)
FTTimeStr = df.loc[0,'InvoiceDate'] #Time at which first transaction of dataset occured
print(FTTimeStr)
FTTime = datetime.strptime(FTTimeStr,'%d/%m/%y %H:%M') #converting from string to datetime format
i = 0
j = 0
df
TimeDiffArr = []
while i<df_size:
    EventTimeStr = df.loc[i,'InvoiceDate']
    EventTime = datetime.strptime(EventTimeStr, '%d/%m/%y %H:%M') #converting from string to datetime format
    TimeDiff = EventTime - FTTime #timedelta format, time difference from first purchase
    TimeDiffVal = (TimeDiff.days*24 + TimeDiff.seconds/3600)/24 #timedelta format converted to days
    TimeDiffArr.append(TimeDiffVal)
    i = i+1
df['Time Difference'] = TimeDiffArr #We replace the InvoiceDate attribute with Time Difference
df = df.drop(columns = ['InvoiceDate'])
df
```

### 2.3.3. Data Transformation - 'Sequence Number'

```python
# In this block we create lists containing all the possible values from 'StockCode'and corresponding values of
#'Description' to create a simple dictionary. 'Description' is then dropped. We then deduce the sequence number for all 
# instances by using 'CustomerID' attribute. A list containing all 'CustomerID' values is obtained using which 'Sequence
# Number' values are accordingly alloted. This attribute is then appended to data frame with the removal of 'CustomerID'

StockCodeSet = [] #List of all 'StockCode' values
DescriptionSet = [] #List of all 'Description' corresponding to 'StockCode'
CustomerIDSet = [] #List of all 'CustomerID' values
i = 0
while i<df_size:
    Code = df.loc[i,'StockCode'] 
    Desc = df.loc[i,'Description'] 
    ID = df.loc[i,'CustomerID']
    if Code not in StockCodeSet: 
        StockCodeSet.append(Code)
        DescriptionSet.append(Desc)
    if ID not in CustomerIDSet:
        CustomerIDSet.append(ID)
    i = i + 1
df = df.drop(columns = ['Description'])

Seq_No = []
i = 0
while i<df_size:
    ID = df.loc[i,'CustomerID']
    Seq_No.append(CustomerIDSet.index(ID)+1)
    i = i+1
df['Sequence Number'] = Seq_No
df = df.sort_values('Sequence Number') #Sorting values based on 'Sequnce Number'
df = df.drop(columns = ['CustomerID'])
df = df.reset_index()
df
```

### 2.3.4. Data Transformation - 'Event Number' 

```python
#The objective of this block is the allocation of 'Event Number' values to all instances. After sorting in the previous
# block with 'Sequence Number', the order of instances in each sequence is also changed. Hence in the first part of this 
# code, sorting is done using 'InvoiceNo' for a temporary data frame containing all instances with a given 'Sequence Nuumber'
# value. The second part involves the allocation of event number based on 'InvoiceNo' values for a given 'Sequence Number'
# value. InvoiceNo is used over Time Difference as the InvoiceNo values are allocated based on time of order. 
i = 0
j = 1 #Refers to the sequence number for which sorting is currently ocurring
prev = 0 #contains value of previously sorted dataframe's sequence number
k = 1
dffin = df.head(0)
while i <df_size:
    SeqNo = df.loc[i,'Sequence Number']
    if SeqNo!= j: #If there is change in sequnce number value, the sorting is done from prev to that value
        dftemp = df[prev:i]
        dftemp = dftemp.sort_values('InvoiceNo')
        dftemp = dftemp.reindex()
        prev = i
        dffin = pd.concat([dffin, dftemp],ignore_index = True) #The sorted dataframe is appened to the final one
        j = j+1
    i = i+1
# The body of loop is repeated one more time to cover all the values of the dataframe
dftemp = df[prev:i]
dftemp = dftemp.sort_values('InvoiceNo')
dftemp = dftemp.reindex()
dffin = pd.concat([dffin, dftemp],ignore_index = True)
dffin.drop(0)
dffin = dffin.drop(columns=['index'])

i = 1
k = 1
Eve_No = [1] #List of event number corresponding to the current arrangement of dataframe
INPrev = dffin.loc[0,'InvoiceNo'] #InvoiceNo values of previous iteration
SNPrev = dffin.loc[0,'Sequence Number'] #SequenceNo values of previous iteration
while i<df_size:
    INCurr = dffin.loc[i,'InvoiceNo']
    SNCurr = dffin.loc[i,'Sequence Number']
    if SNCurr == SNPrev: #While the sequence number of instance equals that in consideration
        if INCurr!=INPrev: #while the invoice number of instance equals that in consideration
            k = k+1
    else:
        k = 1
    Eve_No.append(k)
    INPrev = dffin.loc[i,'InvoiceNo']
    SNPrev = dffin.loc[i,'Sequence Number']
    i = i+1
dffin['Event Number'] = Eve_No    #The 'Event Number' attribute is appended to the dataframe
dffin.head(20)
```

### 2.3.5. Data Transformation - 'ItemSet'

```python
#This is the final step in preprocessing in order to obtain the vertical data format. All items (represented by 'StockCode')
# purchased togther in a particular order (represented by 'Event Number')  by a given customer (represented by 'Sequnce Number')
# are clubbed together i.e. the sequences are obtained. 

dffin2 = pd.DataFrame()
SNo = [] # Updated list of Sequence Numbers after clubbing of items from same order 
ENo = [] # Updated list of Event Numbers after clubbing of items from same order 
IS = [] # Updated list of Item Set after clubbing of items from same order 
TD = []  # Updated list of Time Difference after clubbing of items from same order 
Seq = [] #Temporary variable which will contain the ItemSet generated for each row 
i = 0
SNoPrev = '1'
ENoPrev = '1'
TDPrev = 0
while i<df_size:
    SNoCurr = dffin.loc[i,'Sequence Number']
    ENoCurr = dffin.loc[i,'Event Number']
    ISCurr = dffin.loc[i,'StockCode']
    TDCurr = dffin.loc[i,'Time Difference']
    if SNoCurr == SNoPrev: #If Sequnce Number has not changed
        if ENoCurr==ENoPrev: #If event number has not changed
            Seq.append(ISCurr) # Append the current item to the existing itemset
        else: # If the event numbers have changed then create a new row and restart the itemset sequence
            IS.append(Seq)
            SNo.append(SNoCurr)
            TD.append(TDPrev)
            ENo.append(ENoPrev)
            ENoPrev = ENoCurr
            TDPrev = TDCurr
            Seq = [ISCurr]
    else: # If the sequence numbers have changed then create a new row and restart the itemset sequence
        IS.append(Seq)
        SNo.append(SNoPrev)
        TD.append(TDPrev)
        ENo.append(ENoPrev)
        SNoPrev = SNoCurr
        ENoPrev = 1
        TDPrev = TDCurr
        Seq = [ISCurr]
    i = i + 1
IS.append(Seq)
SNo.append(SNoCurr)
TD.append(TDCurr)
ENo.append(ENoCurr)
dffin2 = pd.DataFrame() #Initilization of new dataframe which will contain the updated obtained values 
dffin2['SequenceNo'] = SNo
dffin2['EventNo'] = ENo
dffin2['ItemSet'] = IS
dffin2['Time Difference'] = TD
dffin2 = dffin2.drop(0)
dffin2 = dffin2.reset_index()
dffin2 = dffin2.drop(columns = ['index'])
dffin2
```

## 4. SPADE Algorithm

### Using the preprocessed dataframe as the initial point, the SPADE algorithm is applied 

### 4.1.1 Determination of all items, their frequency and corresponding positions.

```python
dffin2_size = len(dffin2)
i = 0 
D1 = [] #The overall list of all items present in all order
CountD1 = [] #The corresponding support of each items 
tot = 0
while i<dffin2_size:
    I = copy.deepcopy(dffin2.loc[i,'ItemSet']) #The subsequence for each sequence and event number is taken i.e. 'ItemSet' value
    while I!=[]:
        I1 = I[0] #Each item present in the sequence is accessed sequentially and the count is increased accordingly
        if I1 in D1:
            ind = D1.index(I1)
            CountD1[ind] = CountD1[ind] + 1
        else:
            D1.append(I1)
            CountD1.append(1)
        I.pop(0)
    i = i+1
```

### 4.1.2. Determination of frequent 1-subsequences

```python
l = len(D1)
F1 = [] #list of all frequent 1-subsequences
CountF1 = [] #list of corresponding count of 1-subsequences
min_sup1 = 1000 # minimum support value
i = 0
while i<l:
    if CountD1[i]>min_sup1:
        F1.append(D1[i])
        CountF1.append(CountD1[i])
    i = i+1
```

### 4.1.3. Determination of all position (SeqNo,EveNo) table for each frequent 1-subsequences

```python
i = 0
D = [[]] # This will be a 3D list with each element consisting of a 2-D list with the positions
while i<len(F1):
    D.append([])
    i = i+1
i = 0
while i<dffin2_size:
    I = dffin2.loc[i,'ItemSet'] #Accesing the sequence for a particular sequence and event number
    S = dffin2.loc[i,'SequenceNo']
    E = dffin2.loc[i,'EventNo']
    j = 0
    while j<len(I):
        temp = I[j] #Accesing each element in that sequence
        if temp in F1: #Adding the positions to D if the element is a frequent 1-subsequence
            ind = F1.index(temp)
            D[ind].append([S,E])
        j = j+1
    i = i+1

i = 0
print('List of frequent 1-subsequences and their frequency:')
while i<len(F1):
    print(F1[i]," ",CountF1[i])
    i = i+1
```

### 4.2.1. Generation of 2-subsequence from different frequent 1-subsequences 

```python
#2-subsequeces are generated from different frequent 1-subsequences in this block 
# The COBJ method for support counting is used. 
i = 0
L2 = [] #List of candidate 2-subsequences generated from frequent 1-subsequence in the following loop
CountL2 = []# Corresponding support of 2-subsequences
DD = [[]] #Corresponding position (SeqNo,EveNo) table list of 2-subsequence
while i<len(F1):
    I1 = F1[i] #1st item to compare from frequence 1-subequences
    D1 = D[i]
    Len1 = len(D1)
    j = i+1
    while j<len(F1):
        I2 = F1[j]#2nd item to compare from frequence 1-subequences
        D2 = D[j]
        Len2 = len(D2)
        l = 0
        while l<Len1:
            m = 0
            while m<Len2:
                # Accesing the positions of both items using l and m and comparing using SPADE algorithm rule
                S1 = D1[l][0]
                S2 = D2[m][0]
                if S1==S2: #Checking if sequence number is the same
                    E1 = D1[l][1]
                    E2 = D2[m][1]
                    if E2>E1: #checking ordering of event numbers in order to generate corresponding 2-subsequence
                        ele = [[I1],[I2]] #2-subsequence is [I1],[I2] as I2 is after I1
                        EID1 = E1
                        EID2 = E2
                    elif E1>E2:
                        ele = [[I2],[I1]] #2-subsequence is [I2],[I1] as I1 is after I2
                        EID1 = E2
                        EID2 = E1
                    else:
                        ele = [[I1,I2]] #2-subsequence is [I1,I2] as I1 and I2 have same event number also
                        EID1 = E1
                        EID2 = E2
                    if ele in L2: #If this 2-subsequence has not be encounterd we append it. Else we increase the counter accordingly
                        ind = L2.index(ele)
                        if S1 not in [x[0] for x in DD[ind]]: #It is checked if the 2-subsequence has already been encounterd
                            # for a given sequence number in order to follow COBJ support counting method. 
                            CountL2[ind] = CountL2[ind] +1
                            DD[ind].append([S1,EID1,EID2])
                    else:
                        L2.append(ele)
                        CountL2.append(1)
                        DD.append([])
                        DD[len(DD)-2].append([S1,EID1,EID2])
                m = m+1
            l = l+1
        j = j+1
    i = i+1
```

### 4.2.2. Generation of 2-subsequence from same frequent 1-subsequences 

```python
#Each frequent 1-subsequence must be compared with itself also to generate 2-subsequnce. However in order to ensure accurate
#results, it is done seperately. The previous block compares all frequent 1-subsequence to every other one and the current
##block compares each frequent 1-subsequence to itself. The code structure is the same as that of the previous one. 
i = 0
while i<len(F1):
    I1 = F1[i]
    D1 = D[i]
    Len1 = len(D1)
    l = 0
    while l<Len1:
        m = l+1
        while m<Len1:
            S1 = D1[l][0]
            S2 = D1[m][0]
            if S1==S2:
                E1 = D1[l][1]
                E2 = D1[m][1]
                if E2!=E1:
                    ele = [[I1],[I1]]
                    EID1 = E1
                    EID2 = E2
                else:
                    ele = [[I1,I1]]
                    EID1 = E1
                    EID2 = E2
                if ele in L2:
                    ind = L2.index(ele)
                    if S1 not in [x[0] for x in DD[ind]]:
                        CountL2[ind] = CountL2[ind] +1
                        DD[ind].append([S1,EID1,EID2])
                else:
                    L2.append(ele)
                    CountL2.append(1)
                    DD.append([])
                    DD[len(DD)-2].append([S1,EID1,EID2])
            m = m+1
        l = l+1
    i = i+1
```

### 4.2.3. Determination of frequent 2-subsequences

```python
min_sup2 = 200 #A different value of minimum support is used for 2-subsequences as the thresholds need to be changed 
# based on size in order to achieve good results. 
i = 0 
F2 = [] #List of frequent 2-subsequences
CountF2 = [] #Corresponding support of frequent 2-subsequences
FF = [[]] #Corresponding position table list of frequent 2-subsequences
while i<len(L2):
    if CountL2[i]>=min_sup2:
        F2.append(L2[i])
        CountF2.append(CountL2[i])
        FF.append(DD[i])
    i = i+1
FF.pop(0)    
i = 0
print('Subsequence | Support | 1st 5 occurences in (SeqNo,EveNo1,EveNo2) format')
while i<len(F2):
    print(F2[i]," ",CountF2[i]," ",FF[i][0:5])
    i = i+1
```

### 4.3.1 Generation of frequent n-subsequnces

```python
#Using frequent 2-subsequences as the base case, frequent n-subsequences (n>2) are generated. 
LN = [] #List of n-subsequences generated for given n
CountLN = [] #corresponding count of n-subsequences
DLN = [[]] #corresponding posiitons of n-subsequences
FFin = copy.deepcopy(F2) #Overall list of frequence n-subsequences (n>=2)
FFinCount = copy.deepcopy(CountF2) #Corresponding count of overall list of frequence n-subsequences (n>=2)
FFinD = copy.deepcopy(FF) #Corresponding positions of overall list of frequence n-subsequences (n>=2)
FCurr = copy.deepcopy(F2) #List of n-subsequences used in loop to generate (n+1)-subsequences
FCurrCount = copy.deepcopy(CountF2) #corresponding support of FCurr
FCurrD = copy.deepcopy(FF) #Corresponding positions of FCurr
FNext = [] #List of (n+1)-subequences generated in each iteration
FNextCount = [] #Corresponding support of FNext
FNextD = [[]] #Corresponding positions of FNext
min_sup3 = 100 
while True:
    i = 0
    while i<len(FCurr):
        I1 = FCurr[i] #1st n-subsequence to use for generation
        D1 = FCurrD[i] #Position list of 1st n-subsequence
        Len1 = len(D1) 
        j = i+1 
        while j<len(FCurr):
            I2 = FCurr[j] #2nd n-subsequence to use for generation
            D2 = FCurrD[j] #Position list of 2nd n-subsequence
            Len2 = len(D2)
            Temp1 = copy.deepcopy(I1)
            Temp2 = copy.deepcopy(I2)
            #In order to combine two subsequences, we first need to check if they can be added by deleting the first 
            # element of 1st subsequence and last element of 2nd subsequence
            if len(Temp1[0]) !=1: 
                Temp1[0].pop(0)
            else:
                Temp1.pop(0)
            if len(Temp2[len(Temp2)-1]) !=1:
                Temp2[0].pop(len(Temp2[0])-1)
            else:
                Temp2.pop(len(Temp2)-1)
            if Temp1==Temp2: #We proceed to check position compatibility after this
                l = 0
                while l<Len1:
                    m = 0
                    while m<Len2:
                        #Using the current 2 nested loops we compare all the positions of given n-subsequences. 
                        S1 = D1[l][0]
                        S2 = D2[m][0]
                        Di1 = copy.deepcopy(D1[l])
                        Di2 = copy.deepcopy(D2[m])
                        if S1==S2:#First we check for sequence number compatibility
                            Temp3 = copy.deepcopy(Di1)
                            Temp4 = copy.deepcopy(Di2)
                            Temp3.pop(0)
                            Temp4.pop(0)
                            Temp3.pop(0)
                            Temp4.pop(len(Temp4)-1)
                            if Temp3==Temp4: #Secondly we check if the event numbers of these 2 subsequences match by 
                                # removing 1st event number of 1st subsequence and last of second subsequence. 
                                Temp5 = copy.deepcopy(I2)
                                add = Temp5[len(Temp5)-1]
                                if len(add)>1:
                                    addfin = add[len(add)-1]
                                else:
                                    addfin = add[0]
                                #addfin is the element to be added to the 1st subsequence in order to generate the 
                                #(n+1) subsequence
                                ele = copy.deepcopy(I1)
                                T1 = copy.deepcopy(Di1)
                                T2 = copy.deepcopy(Di2)
                                if T1[len(T1)-1]==T2[len(T2)-1]:
                                    ele[len(ele)-1].append(addfin)
                                else:
                                    ele.append([addfin])
                                S = copy.deepcopy(S1) #S is the sequnce number to be appended for the new subsequence
                                E = copy.deepcopy(Di1) #E is the series of event numbers to be appended for the new subsequnce
                                add2 = copy.deepcopy(Di2[len(Di2)-1])
                                E.append(add2)
                                if ele in FNext: #If the subsequence has already been generated then only support and position
                                    #table is updated. Else the subsequence is also appeneded. 
                                    ind = FNext.index(ele)
                                    if S not in [x[0] for x in FNextD[ind]]:
                                        FNextCount[ind] = FNextCount[ind] +1
                                        FNextD[ind].append(E)
                                else:
                                    FNext.append(ele) #FNext contains all (n+1)-subsequences generated
                                    FNextCount.append(1)
                                    FNextD.append([])
                                    FNextD[len(FNextD)-2].append(E)
                        m = m+1
                    l = l+1
            j = j+1
        i = i+1
    p = 0
    FCurr = []
    FCurrCount = []
    FCurrD = [[]]
    while p<len(FNext):
        if FNextCount[p]>=min_sup3: #if the subsequences satisfy the minimum threshold, then they are appended to the 
            # final frequent list (FFin) and the list to be used as base for next iteration (FCurr)
            FFin.append(copy.deepcopy(FNext[p]))
            FFinCount.append(copy.deepcopy(FNextCount[p]))
            FFinD.append(copy.deepcopy(FNextD[p]))
            FCurr.append(copy.deepcopy(FNext[p]))
            FCurrCount.append(copy.deepcopy(FNextCount[p]))
            FCurrD.append(copy.deepcopy(FNextD[p]))
        p = p+1
    FNext = []
    FNextCount = []
    FNextD = []
    if FCurr==[]: #If no new frequent subsequences are generated, then the loop is exited 
        break
```

### 4.3.2 Printing all frequent subsequences
